Import the required libraries and modules that you would need.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score,recall_score,f1_score
from sklearn.utils import resample
```

Read that data into Python and call the dataframe donors.

We make use of the cleaned cleaned version of learningSet.csv from the class - in form of numerical.csv, categorical.csv and target.csv

```python
numerical=pd.read_csv("files_for_lab/numerical.csv")
categorical=pd.read_csv("files_for_lab/categorical.csv")
target=pd.read_csv("files_for_lab/target.csv")
display(numerical.head())
display(categorical.head())
display(target.head())
```

Check for NULLs in categorical

```python
pd.DataFrame(categorical.isna().sum()/len(categorical)).head(60)
```

```python
categorical['FIRSTDATE_MM'].value_counts(dropna=False)
```

```python
indexes_removal=categorical[categorical['FIRSTDATE_MM'].isna()].index
numerical=numerical[~numerical.index.isin(indexes_removal)]
categorical=categorical[~categorical.index.isin(indexes_removal)]
target=target[~target.index.isin(indexes_removal)]
display(numerical.shape)
display(categorical.shape)
display(target.shape)
```

```python
display(categorical['FIRSTDATE_MM'].value_counts(dropna=False))
```

Concatenate numerical and categorical back together again for your X dataframe. Designate the Target as y.

```python
X=pd.concat([numerical,categorical],axis=1)
X.head()
```

```python
y=target['TARGET_B']
y.head()
```

Split the data into a training set and a test set.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1254)
```

Split further into train_num and train_cat. Also test_num and test_cat.

```python
X_train_num = X_train.select_dtypes(np.number)
X_test_num = X_test.select_dtypes(np.number)

X_train_cat = X_train.select_dtypes(object)
X_test_cat = X_test.select_dtypes(object)
```

Scale the features either by using normalizer or a standard scaler. (train_num, test_num)

```python
transformer=MinMaxScaler().fit(X_train_num)
X_train_num_scaled = transformer.transform(X_train_num)
X_train_num_scaled = pd.DataFrame(X_train_num_scaled,columns=X_train_num.columns)
X_train_num_scaled.head()
```

```python
X_test_num_scaled = transformer.transform(X_test_num)
X_test_num_scaled = pd.DataFrame(X_test_num_scaled,columns=X_test_num.columns)
X_test_num_scaled.head()
```

Encode the categorical features using One-Hot Encoding or Ordinal Encoding. (train_cat, test_cat)

```python
encoder = OneHotEncoder().fit(X_train_cat)
X_train_cat_encoded = encoder.transform(X_train_cat).toarray()
cols = encoder.get_feature_names_out(input_features=X_train_cat.columns)
X_train_cat_encoded = pd.DataFrame(X_train_cat_encoded, columns=cols)
X_train_cat_encoded.head()
```

```python
X_test_cat_encoded = encoder.transform(X_test_cat).toarray()
cols = encoder.get_feature_names_out(input_features=X_test_cat.columns)
X_test_cat_encoded = pd.DataFrame(X_test_cat_encoded, columns=cols)
X_test_cat_encoded.head()
```

Re-concatenate train_num and train_cat as X_train as well as test_num and test_cat as X_test

```python
X_train = pd.concat([X_train_num_scaled, X_train_cat_encoded], axis=1)
X_train.head()
```

```python
X_test = pd.concat([X_test_num_scaled, X_test_cat_encoded], axis=1)
X_test.head()
```

Fit a logistic regression model on the training data.

```python
classification = LogisticRegression(random_state=1122, solver='lbfgs')
classification.fit(X_train,y_train)
```

```python
predictions = classification.predict(X_train)
print("Training score: ",round(classification.score(X_train, y_train),2),"%")

predictions = classification.predict(X_test)
print("Test score: ",round(classification.score(X_test, y_test),2),"%")
```

```python
confusion_matrix(y_test, predictions)
```

```python
print("precision: ",precision_score(y_test,predictions))
print("recall: ",recall_score(y_test,predictions))
print("f1: ",f1_score(y_test,predictions))
```

Recall is just 54% which is very low. It means that only 54% of the donors can be predicted from the total number of actual donors, even though we got high accuracy probably due to high number of non-donors, resulting in biasness of non donor prediction.

## Managing imbalance in the dataset

Check for the imbalance.

```python
train_set=pd.concat([X_train,y_train.reset_index(drop=True)],axis=1)
display(train_set.shape)
train_set.head()
```

```python
train_set['TARGET_B'].value_counts()
```

```python
category_0 = train_set[train_set['TARGET_B'] == 0]
category_1 = train_set[train_set['TARGET_B'] == 1]
```

## Downsampling (undersampling)

```python
category_0_undersampled = resample(category_0, 
                                   replace=False, 
                                   n_samples = len(category_1))
```

```python
print(category_0_undersampled.shape)
print(category_1.shape)
```

```python
df_donors_downsampled = pd.concat([category_0_undersampled, category_1], axis=0)
df_donors_downsampled = df_donors_downsampled.sample(frac=1)
display(df_donors_downsampled['TARGET_B'].value_counts())
df_donors_downsampled.head()
```

Build the Logisitic Regression with undersampled Training Data

```python
X_train=df_donors_downsampled.drop(['TARGET_B'],axis=1)
y_train=df_donors_downsampled['TARGET_B']

classification = LogisticRegression(random_state=1122, solver='lbfgs')
classification.fit(X_train,y_train)
```

```python
predictions = classification.predict(X_train)
print("Training score: ",round(classification.score(X_train, y_train),2),"%")

predictions = classification.predict(X_test)
print("Test score: ",round(classification.score(X_test, y_test),2),"%")
```

```python
confusion_matrix(y_test, predictions)
```

```python
print("precision: ",precision_score(y_test,predictions))
print("recall: ",recall_score(y_test,predictions))
print("f1: ",f1_score(y_test,predictions))
```

As we observe, with downsampling, the accuracy has significantly gone down, possibly due to elimination of rows, which ofcourse removes meaningful data from the dataframe.

Regarding the recall, there has been a slight improvement in the value, probably due to the downsampling.

## Upsampling (Oversampling)

```python
category_1_oversampled = resample(category_1, 
                                  replace=True, 
                                  n_samples = len(category_0))
```

```python
print(category_0.shape)
print(category_1_oversampled.shape)
```

```python
df_donors_upsampled = pd.concat([category_0, category_1_oversampled], axis=0)
df_donors_upsampled = df_donors_upsampled.sample(frac=1)
display(df_donors_upsampled['TARGET_B'].value_counts())
df_donors_upsampled.head()
```

```python
X_train=df_donors_upsampled.drop(['TARGET_B'],axis=1)
y_train=df_donors_upsampled['TARGET_B']

classification = LogisticRegression(random_state=1122, solver='lbfgs')
classification.fit(X_train,y_train)
```

```python
predictions = classification.predict(X_train)
print("Training score: ",round(classification.score(X_train, y_train),2),"%")

predictions = classification.predict(X_test)
print("Test score: ",round(classification.score(X_test, y_test),2),"%")
```

```python
confusion_matrix(y_test, predictions)
```

```python
print("precision: ",precision_score(y_test,predictions))
print("recall: ",recall_score(y_test,predictions))
print("f1: ",f1_score(y_test,predictions))
```

As we observe, with upsampling, the accuracy improved a bit as compared to downsampling but at the cost of a slight lower recall, which is nearly the same as calculated with the model without any resampling.

In all the downsampling performed the best in terms of recall even though it costed the accuracy, but for us, we need to identify the right donors and recall is an important metric for that.
